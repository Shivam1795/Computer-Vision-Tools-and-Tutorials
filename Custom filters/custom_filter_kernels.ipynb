```python
## Import all the required libraries ##
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Read image !!
image = cv2.imread('images/flower.jpg')

## Convert image to RGB Colorspace !!
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

## Create a grayscale copy of image !!
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

## Display images !!
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
ax1.imshow(image); ax1.axis(False);ax1.title.set_text('Original RGB Image')
ax2.imshow(gray, cmap='gray'); ax2.axis(False);ax2.title.set_text('Gray scale Image')
```

```python
## Define custom LPF kernels (For blur effect) !!

## Averaging filter !!
avg_kernel = np.ones((3,3))

## Normalized Averaging filter kernel !!
avg_filter = np.multiply(avg_kernel, 1.0/np.sum(avg_kernel))

## Gaussian Blur filter !!
gaussian_kernel = np.array([[1, 2, 1],
                            [2, 4, 2],
                            [1, 2, 1]])

## Normalized Gaussian filter kernel !!
gaussian_filter = np.multiply(gaussian_kernel, 1.0/np.sum(gaussian_kernel))
```

```python
## Apply filters to filter out high frequency noise (custom and openCV inbuild Gaussian Blur filter)!!
## Perform convolution using filter2D, input: gray_image, bit-depth, filter_kernel !!

avg_filtered_image = cv2.filter2D(gray, -1, avg_filter)
gaussian_filtered_img = cv2.filter2D(gray, -1, gaussian_filter)

## Inbuild GaussianBlur filter (Input: Gray_image, kernel_size, variance) !!
gaussianBlured_image = cv2.GaussianBlur(gray, (5,5), 0)

## Display filtered images !!
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.imshow(avg_filtered_image, cmap='gray'); ax1.axis(False); ax1.title.set_text('Average filtered image')
ax2.imshow(gaussian_filtered_img, cmap='gray');ax2.axis(False); ax2.title.set_text('Gaussian filtered image')
ax3.imshow(gaussianBlured_image, cmap='gray');ax3.axis(False); ax3.title.set_text('GaussianBlured image');
```

```python
## Define custom HPF kernels !!

## Sobel_x (For vertical edges) !!
SobelX = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

## Sobel_y (For Horizontal edges) !!
SobelY = np.array([[-1, -2, -1],
                   [ 0,  0,  0],
                   [ 1,  2,  1]])

## Laplacian edge filter !!
laplacian=np.array([[0, 2, 0],
                    [2,-8, 2],
                    [0, 2, 0]])
```

```python
## Apply filters to detect edges !!
## Perform convolution using filter2D, input: gray_image, bit-depth, filter_kernel !!

vertical_edges = cv2.filter2D(gray, -1, SobelX)
horizontal_edges = cv2.filter2D(gray, -1, SobelY)
laplacian_edges = cv2.filter2D(gray, -1, laplacian)

## Display filtered images !!
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.imshow(vertical_edges, cmap='gray'); ax1.axis(False); ax1.title.set_text('Detected vertical edges')
ax2.imshow(horizontal_edges, cmap='gray');ax2.axis(False); ax2.title.set_text('Deteted horizontal edges')
ax3.imshow(laplacian_edges, cmap='gray');ax3.axis(False); ax3.title.set_text('Laplacian edges');
```
