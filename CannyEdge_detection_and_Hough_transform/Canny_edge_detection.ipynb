```python
## Import libraries !!
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
## Read image !!
image = cv2.imread('images/flower.jpg')

## Convert image to RGB from BGR !!
imgRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

## Create a grayscale copy of image for processing !!
imgGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

## Display images !!
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.imshow(imgRGB), ax1.axis(False), ax1.set_title('Original RGB image')
ax2.imshow(imgGray, cmap='gray'), ax2.axis(False), ax2.set_title('Gray scale image');
```

```python
## Try Canny using "wide" and "tight" thresholds !!

wide = cv2.Canny(imgGray, 30, 200)
tight = cv2.Canny(imgGray, 200, 240)
 
    
## Display the images !!
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.imshow(wide, cmap='gray'), ax1.set_title('wide'), ax1.axis(False)
ax2.imshow(tight, cmap='gray'), ax2.set_title('tight'), ax2.axis(False);
```
